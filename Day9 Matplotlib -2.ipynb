```python
def f(t):
    return (np.exp(-t)*np.cos(2*np.pi*t))

t1=np.arange(10,20,2)
t2= np.arange (11,21,2)
plt.subplot (221)
plt.plot(t1,f(t1),'bo',t2,f(t2))
plt.title ('Graph 1')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')


plt.subplot(222)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.title ('Graph 2')
plt.xlabel ('X-Axis')
plt.ylabel ('Y-Axis')

plt.show()
```
